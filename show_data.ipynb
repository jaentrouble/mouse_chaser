```python
import pickle
import matplotlib.pyplot as plt
from skimage import draw
```

```python
%matplotlib inline
```

```python
with open('records/save/0.pck','rb') as f:
    datum = pickle.load(f)
```

```python
food_n=[]
for sample_datum in datum:
    image = sample_datum['image'].copy()
    food = sample_datum['food']
    food_n.append(len(food))
```

```python
plt.plot(food_n)
plt.show()
```

```python
xx, yy = draw.disk(sample_datum['nose'],5)
image[xx,yy] = [255,255,0]
plt.imshow(image)
```
