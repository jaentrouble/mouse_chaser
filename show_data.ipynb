```python
import pickle
import matplotlib.pyplot as plt
from skimage import draw
```

```python
%matplotlib inline
```

```python
with open('records/save/0.pck','rb') as f:
    datum = pickle.load(f)
```

```python
sample_datum = datum[100]
```

```python
image = sample_datum['image']
nose = sample_datum['nose']
```

```python
rr, cc = draw.disk(nose,10)
image[rr, cc] = [0,255,0]
plt.figure(figsize=(8,8))
plt.imshow(image)
```
